## 1) For each *university*, extract from the *times* dataset the most recent and the least recent data, obtaining two separate dataframes


```python
import pandas as pd
import re
import numpy as np
```

### First we imported the *times* dataset

```python
times = pd.read_csv("https://raw.githubusercontent.com/Vitorg-ux/Project2020/main/timesData.csv", engine='python', na_values = '-',thousands =',')
times.head()
```

```python
#Perchè abbiamo messo "na_values = '-' " sopra?
```

```python
most_recent = pd.DataFrame()
for universita in times['university_name'].unique():
    most_recent = most_recent.append(times.loc[times[times['university_name'] == universita]['year'].idxmax()], ignore_index = True)
most_recent.tail()
```

```python
least_recent = pd.DataFrame()
for universita in times['university_name'].unique():
    least_recent = least_recent.append(times.loc[times[times['university_name'] == universita]['year'].idxmin()], ignore_index = True)
least_recent.tail()
```

## 2) For each *university*, compute the improvement in income between the *least* recent and the *most* recent data points

```python
uniti = pd.merge(most_recent, least_recent, on = 'university_name')
uniti.head()
```

```python
uniti['improvement'] = pd.to_numeric(uniti['income_x']) - pd.to_numeric(uniti['income_y'])
uniti.head()
```

## 3) Find the university with the *largest increase* computed in the previous point

```python
uniti[uniti['improvement'] >= uniti['improvement'].max()][['university_name','improvement']]
```

## 5) Consider only the most recent data point of the *times dataset*. Compute the number of male and female students for each country.¶

```python
mr_times = most_recent.dropna(subset= ['female_male_ratio', 'num_students']).copy()
mr_times.tail()
```

```python
appoggio = mr_times['female_male_ratio'].str.extract(r'(\d+) : (\d+)')
mr_times['females'] = mr_times['num_students']*appoggio[0].astype(int)//(appoggio[0].astype(int)+appoggio[1].astype(int))
mr_times.head()          #se puoi ottimizza!!!
```

```python
mr_times['males'] = mr_times['num_students']*appoggio[1].astype(int)//(appoggio[1].astype(int)+appoggio[0].astype(int))
mr_times
```

```python
mr_times.groupby('country')[['country', 'females', 'males']].sum()
```

## 6) Find the universities where the *ratio* between female and male is below the average ratio (computed over all universities)

```python
mr_times['ratio'] = mr_times['females']/mr_times['males']
mr_times.head()
```

```python
#mr_times[mr_times['males']!= 0]['ratio'].mean()
```

```python
average_ratio = mr_times['females'].sum()/mr_times['males'].sum()
average_ratio
#qui abbiamo calcolato il totale di donne diviso il totale degli uomini
```

```python
punto5 = mr_times[mr_times['ratio']< average_ratio]
punto5.head()
```

```python
punto5.sort_values(by=['ratio']).tail()      #giusto per essere sicuri sicuri
```
